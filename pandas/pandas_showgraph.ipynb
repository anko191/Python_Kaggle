```
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt 
```

```
from sklearn import datasets
iris = datasets.load_iris()
```

```
df = pd.read_csv('./iris.csv', index_col = None)
print(df.head())
```

```
df.plot()
```

## plt.savefig()
## plt.show()
## plt.figure() reset !

## plt.close('all') !

```
plt.figure()
df.plot()
plt.savefig('./pandas_iris_line.png')
plt.close('all')
```

```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
df.plot(ax = ax)
fig.savefig('./pandas_iris.png')
```

```
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize=(9,6))
df.plot(ax = axes[0,0], legend = False)
df.plot(ax=axes[1,2], legend = False, kind = 'hist')
fig.savefig('./na_iris.png')
```

# setting argument
* title
* grid
* colormap
* legend
* alpha : dont show

```
df.plot(title = 'Iris Data Set',
        grid = True,
        colormap = 'Accent',
        legend = False,
        alpha = 0.5)
```

## argument 'kind':
* line : line plot
* bar : vertical bar plot
* barh : horizontal bar plot
* box : boxplot
* hist : histogram
* kde, density : Kernel Density Estimation plot
* area : area plot
* scatter : scatter plot
* hexbin : hexbin plot
* pie : pie plot

```
df.plot(kind='line')
```
