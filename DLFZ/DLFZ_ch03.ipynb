```
# sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.grid()
```

```
y1 = sigmoid(x)
y2 = step_function(x)
plt.plot(x,y1, label="sigmoid")
plt.plot(x,y2, linestyle = "--", label="step_function")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sigmoid & step_function")
plt.legend()
plt.grid()
```

```
# relu
def relu(x):
    return np.maximum(0, x)
```

```
y3 = relu(x)
```
