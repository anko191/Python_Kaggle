```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

ref : https://www.geeksforgeeks.org/image-segmentation-using-k-means-clustering/

```python
def kmeans_image(name, k = 3):
    
    image = cv2.imread(name)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # plt.figure(figsize=(5,10))
    fig,ax = plt.subplots(1,2, figsize = (12,24))
    fig.patch.set_facecolor('white')
    ax[0].imshow(image)
    ax[0].set_axis_off()
    ax[0].set_title('Before')
    
    pixel_vals = image.reshape((-1,3))
    pixel_vals = np.float32(pixel_vals)
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,
                100, 0.85)
    retval, labels, centers = cv2.kmeans(pixel_vals,
                                         k, None,
                                         criteria,
                                         10,
                                         cv2.KMEANS_RANDOM_CENTERS)
    centers = np.uint8(centers)
    segmented_data = centers[labels.flatten()]
    segmented_image = segmented_data.reshape((image.shape))
    ax[1].imshow(segmented_image)
    ax[1].set_axis_off()
    ax[1].set_title(f'After(k={k})')
    
```

```python
kmeans_image('blessyou.png', k = 5)
```

```python
kmeans_image('wingman.png', k = 3)
```

```python
kmeans_image('champ.png', k = 5)
```
