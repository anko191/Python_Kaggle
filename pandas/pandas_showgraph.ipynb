# heatmap with Seaborn

```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## seaborn.heatmap()

```
list_2d = [[0,1,2],[3,4,5]]
sns.heatmap(list_2d)
```

```
plt.figure() # ~
sns.heatmap(list_2d)
plt.savefig('./l.png')
plt.close('all')
```

```
arr_2d = np.arange(-8,8).reshape((4,4))
print(arr_2d)
```

```
plt.figure()
sns.heatmap(arr_2d)
```

## in case 'pandas.DataFrame'

```
df = pd.DataFrame(data = arr_2d, index = ['a','b','c','d'], columns=['A','B','C','D'])
print(df)
```

```
plt.figure()
sns.heatmap(df)
```

### return AxesSubplot

```
print(type(sns.heatmap(list_2d)))
```

```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.heatmap(list_2d, ax = ax)
```

```
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize=(8,6))
sns.heatmap(list_2d, ax = axes[0,0])
sns.heatmap(arr_2d, ax=axes[1,2])
```

# seaborn.heatmap()'s arguments

## 'annot = True' show numerics in heatmap

```
sns.heatmap(df, annot = True)
```

## 'cbar = False' dont show colorbar in heatmap

```
sns.heatmap(df, cbar=False)
```

## 'square = True' show in square in heatmap

```
sns.heatmap(df, square = True)
```

## 'vmax = ?, vmin = ?, center = ?' settings

```
sns.heatmap(df, vmax = 10, vmin = -10, center = 0)
```

## 'cmap = hot etc...' designate colormap in heatmap

```
sns.heatmap(df, cmap = 'hot')
```

## if you 'cmap = ?' adding '_r', color reversing.

```
sns.heatmap(df, cmap='Blues')
```

```
sns.heatmap(df, cmap='Blues_r')
```

## size settings

```
current_figsize = mpl.rcParams['figure.figsize']
print(current_figsize)
```

```
plt.figure(figsize=(9,6))
sns.heatmap(df, square = True)
```

```
current_dpi = mpl.rcParams['figure.dpi']
print(current_dpi)
plt.figure()
sns.heatmap(df, square = True)
```

# pairplot() with seaborn

```
import pandas as pd
import seaborn as sns
df = pd.read_csv('./iris.csv', index_col = False)
# df = sns.load_dataset("iris")
```

```
print(df.head())
print(df.dtypes)
```

```
print(df['species'].value_counts())
```

## return seaborn.PairGrid class

```
pg = sns.pairplot(df)
print(type(pg))
```

## Color coded according to category data

```
sns.pairplot(df, hue='species')
```

## category order 'hue_order'

```
sns.pairplot(df, hue='species', hue_order = ['virginica', 'versicolor', 'setosa'])
```

```
sns.pairplot(df, hue = 'species', palette = 'Blues')
```

```
sns.pairplot(df, hue='species', palette={'setosa':'red',
                                         'versicolor':'#00ff00',
                                         'virginica':'blue'})
```

## vars, x, y

```
sns.pairplot(df, hue='species', vars = ['sepal_length', 'sepal_width'])
```

## x_vars, y_vars

```
sns.pairplot(df, hue='species',
             x_vars = ['sepal_length', 'sepal_width'],
             y_vars = ['petal_length', 'petal_width'])
```

## designate markers

```
sns.pairplot(df, hue='species', markers='+')
```

```
sns.pairplot(df, hue='species', markers=['+','s','d'])
```

## if kind = 'reg'..., Regression Line

```
sns.pairplot(df, hue='species', kind = 'reg')
```

## argument diag_kind = 'kde'

```
sns.pairplot(df, hue='species', diag_kind='kde')
```

```
sns.pairplot(df, hue='species', diag_kind='hist')
```

```
sns.pairplot(df, hue='species', size = 2, diag_kind = 'hist')
```

## plot_kws('alpha':0.2), diag_kws('bins':20,'histtype':'step')

```
sns.pairplot(df, hue='species',
             plot_kws={'alpha':0.2},
             diag_kws={'bins':20,'histtype':'step'},diag_kind = 'hist')
```

```
sns.pairplot(df, hue='species', kind='reg',
             plot_kws={'ci': None,
                       'marker': '+',
                       'scatter_kws': {'alpha': 0.4},
                       'line_kws': {'linestyle': '--'}},diag_kind = 'hist')
```
