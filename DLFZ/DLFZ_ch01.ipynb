```
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    def hello(self):
        print("Hello "+self.name+"!")
    def goodbye(self):
        print("Good-bye "+self.name+"!")
m = Man("David")
m.hello()
m.goodbye()
```

```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,6,0.1)
y = np.sin(x)
```

```
plt.plot(x,y)
plt.grid()
```

```
y1 = np.sin(x)
y2 = np.cos(x)
```

```
plt.plot(x,y1,label="sin")
plt.plot(x,y2,label="cos", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.grid()
```
